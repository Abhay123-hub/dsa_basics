```python
## the above code solution which i have written is correct
## but that is giving us time complaexity of O(n^2) but we need to get time complexity of O(n)
## for this we need to use single loop only not the nested loop
## so how i can solve this problem by using single loop
## since there is a condition that list 'm' elements can be only in the range of [1,10]
## so now first i will define a list of length(11) indexing from 0 to 10
## this list will be initialized such that all elements will be zero
## and it will be used to store the frequecy of elements of the list m
## such that if in the third index there is 2 it means 3 has come two times in the list 'm'
```

```python
m = [1,2,2,3,4,4,5,7,5,8,9,10,9,10,10]
n = [11,2,8,9,7,6,90,2]
hashed_list = [0]*11
dict = {}

for num in m:
    hashed_list[num]+=1
## now how will i use this hashed list to find the count of elements of n in the list m
## let us see how the things will work here
for num in n:
    ## of the number is out of the range of list m
    ## then there is no point to consider that number
    ## only consider of the number is in the range of list m
    
    if num<1 or num >10:
        dict[num] = 0
    else:
        dict[num] = hashed_list[num]

print(dict)
        
## for this solution the time complexity is O(m+n)
## for worst case O(2n) which becomes O(n) 
## so here we have obatined the required time complexity and tackled the time limit exceeding error
```
