```python
## this code is for sorting in ascending order
def sort(arr):
    for i in range(len(arr)-1,0,-1):
        for j in range(i):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]
    return arr
```

```python
sort([1,3,2,7,4]) ## time comlexity (n*2) taking the worst case scienerio for the time complexity 
```
