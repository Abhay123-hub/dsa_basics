```python
## sorting the array using bubble sort
## so what is the intution behind the bubble sort method
## here you compare two adjacent elements 
## if the first element is greater then the second element, then swap these two elements otherwise do not swap
## this algorithm will work such that after each i iteration largest element will be at last of the array
## now let us write the code of bubble sort methpd for sorting the array in ascending order
```

```python
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(0,len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]
    return arr
```

```python
bubble_sort([4,2,1,3,8,18])
```

```python
## now i will be sorting the array using insertion sort
## so how actually this method works
## what is the intution behind the insertion sort method?
## first i will iterate through the array with first loop
## first i will check with first two adjacent elements of the array
## if the first element is lesser than the second element then it means the array is sorted till second index
## now in next  we will also check the for next two adjacent elements the first element is greater than the second element
## if it is true then it means till third index all the elements are sorted in the given array
## now  next i we will do the same if our given condition is not satisfied it means till fourth element our array 
## is not sorted , so now with respect to the fourth element we need to sort the array
## now there are multiple case possible the correct position of the fourth can be any till third position of the array
## of course it is not going to be the fourth position
## so first of all i will store the fourth index in a variable and will replace the fourth element with the third element
## why i am doing this we will able to know it next
## now at both the third and fourth position we will have third element 
## now we will iterate a reverse loop from the third element to zero element
## see if the fourth element stored in variable is less than the particular element
## now if the fourth element is less than second element then in third and second position you will add the second element
## and now again for loop will move forward we will see if the first element is less than fourth if not it means
## fourth element should be in second position so i will put variable in the second position
## now after this till fourth element the array is sorted
## can i write the code just based on this concept whiich will actual;y sort the array using insertion sort method
```

```python
## let us writting the code for insertion sort for sorting the array
## by keeping all the things in mind
```

```python
def insertion_sort(arr):
    for i in range(len(arr)-1):
        if arr[i+1] < arr[i]: ## at this case the array is not sorted properly
            ## means arr[i] is not in  its correct position i need to put into its correct position
            ## store arr[i] in a variable named key
            key = arr[i+1]
            ## now replace the arr[i+1] by arr[i]
            ## such that at both index i and i+1 , there will be element arr[i]
            arr[i+1] = arr[i]
            ## now i will reverse back from i -1 to including zero index
            ## and will try to find the correct position of the key element in the provided array
            for j in range(i-1,-2,-1):
                if j == -1:
                    arr[j+1] = key
                elif arr[j] > key: 
                    arr[j+1] = arr[j]
                else:
                    arr[j+1] = key
                    break
    return arr         
```
