```python
## print "hello world" four times using recursion
count = 0
def hello():
    global count
    if count==4:
        return 
    print("hello world")
    count+=1
    hello()
    
    
```

```python
hello()## this was the example of head recursion
## head recursion means first doing the task then calling the function inside the function
```

```python
## now we will do the example of tail recursion
## as we know that in the head recursion first the task is performed then function is called
## and in tail recursion first the function is called then task is performed ,the tail recursion is also called back tracking
```

```python
## example of tail recursion with same example
count = 0
def hello():
    global count
    if count == 4:
        return
    count+=1
    hello()
    print("hello world")
```

```python
hello()
```

```python
def reverse_array(lst, new_lst, i):
    if i == len(lst):   # stop when we reach the end of lst
        return new_lst
    reverse_array(lst, new_lst, i + 1)  # recurse till end
    new_lst.append(lst[i])              # append in reverse order
    return new_lst

# Example
arr = [1, 2, 3, 4, 5]
print(reverse_array(arr, [], 0))  # [5, 4, 3, 2, 1]
```

```python
##code for reversing the array using while loop
def reverse(arr):
    left = 0
    right = len(arr) - 1
    while left < right:
        arr[left],arr[right] = arr[right],arr[left]
        left+=1
        right-=1
    return arr
```
